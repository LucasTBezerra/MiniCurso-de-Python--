```python
# Mini Curso de Python 

# importar a base de dados
# Percorrer a base de dados para verificar se há possíveis cagadas nas linhas e nas colunas 
# calcular o faturamento por loja 
# calcular a quantidade de produtos vendidos por loja 
# ticket médio por produto em cada loja 

# Enviar um e-mail com um relatório final 

import pandas as pd 
tabela = pd.read_excel('Vendas.xlsx')
display(tabela)

# Caso todas as colunas não sejam mostradas USE -->>>  pd.set_option('display.max_coluns', None)
```

```python
tabela = tabela.drop('Código Venda', axis=1)
tabela = tabela.drop('Data', axis=1)
tabela = tabela.dropna()
display(tabela)
```

```python
faturamento_total = tabela['Valor Final'].sum()
print(faturamento_total)

# Faturamento por loja            
# Se vamos analisar uma coluna só basta 1 [], se são mais de 1 em diante.. [[]] dois colchetes 
faturamento_p_loja = tabela[['ID Loja', 'Valor Final']] 
display(faturamento_p_loja)

# Seria melhor agrupar essas lojas somando o faturamento de cada uma. 

faturamento_p_loja = tabela[['ID Loja', 'Valor Final']].groupby('ID Loja').sum() # da mesma forma q podemos pegar o total sum poderiamos pegar a média ou etc..
display(faturamento_p_loja)
```

```python
# A principio não iriamos por [] após o groupby mas como estamos tratando de informações uma uma lista colocamos ([])

faturamento_p_produto = tabela[['ID Loja', 'Produto', 'Quantidade', 'Valor Final']].groupby(['ID Loja', 'Produto']).sum() 
display(faturamento_p_produto)
```

```python
# ticket médio por produto em cada loja.  
# O ticket é a quantidade de vendas total pelo número de produtos vendidos

# Sempre que fazemos uma operação onde haverá uma coluna dividia pela outra aparecerá um erro - dtype: float64 - pq a resposta n necessariamente será uma tabla. 
# Para resolvermo isso colocamos a operação entre    ( ssdbstbstetehethth ).toframe()      então será convertido em tabela. 

ticket_med = (faturamento_p_loja['Valor Final']/faturamento_p_produto['Quantidade']).to_frame()
ticket_med = ticket_med.rename(columns={0: 'Ticket Medio'})
display(ticket_med)
```

```python
# Agora vamos enviar um e-mail com os dados em forma de relatório

# import win32com.client as win32
# outlook = win32.Dispatch('outlook.application')
# mail = outlook.CreateItem(0)
# mail.To = 'To address'
# mail.Subject = 'Message subject'
# mail.HTMLBody = '<h2>HTML Message body</h2>'
# mail.send()

# quando temos que escrever um texto com mais de 1 linha devemos escreve-lo com 3 aspas simples     ''' '''
# Precisamos tbm editar o meail para que ele tenha parágrafos, e um corpo estruturado 
# Em texto, todo parágrafo esta entre   <p> texto do parágrafo </p>
# como vamos precisar de que as tabelas sejam introduzidas ao longo do texto, vamos adicionar o f antes das '''
# Agora podemos introduzir dentro das {} a variável que queremos ..
# colocaremos .to_html() para converter as tabelas em uma tabrla exibível no e-mail. 

# Um outro detalhe que precisamos resolvere é o separador de casas decimais para que fique de mais facil compreensão 
# para isso, usaremos ... formatters={'Valor Final': 'R${:,.2f}'.format}
# Usamos 'Valor Final': pq é nessa coluna que iremos alterar o valor para  : R${:,.2f} o 2 f é pa são duas casas decimais, caso fosse apenas uma 1f


import win32com.client as win32
outlook = win32.Dispatch('outlook.application')
mail = outlook.CreateItem(0)
mail.To = 'lucastelesbezerra@hotmail.com'
mail.Subject = 'Relatório de Dados'
mail.HTMLBody = f''' 
<p> Prezados,</p>

<p> Segue em anexo o relatório contendo um resumo dos dados do arquivo 'Vendas.xlsx' tratdos por python, contendo:</p> 
<p> Total de vendas por loja,</p> 
<p> Total de vendas por produto,</p> 
<p> Ticket por produto.</p> 

<p> Faturamento por Loja</p> 
{faturamento_p_loja.to_html(formatters={'Valor Final': 'R${:,.2f}'.format})}

<p> Faturamento por produto</p> 
{faturamento_p_produto.to_html()}

<p> Ticket médio por produto</p>
{ticket_med.to_html(formatters={'Ticket Médio': 'R${:,.2f}'.format})}

<p> Qualquer dúvida estou a disposição,</p>

<p>Lucas T. Bezerra</p>
'''
mail.send()
print('email enviado')
```
